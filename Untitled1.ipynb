# Compare Z3D and Observatory data

Assuming that a MTH5 has been made from a ZEN station and USGS geomagnetic data.  We just want to plot the data to make sure it looks correct.

```python
from mth5.mth5 import MTH5
```

```python
%matplotlib widget
```

```python
m_zen = MTH5()
m_zen.open_mth5(r"c:\Users\jpeacock\OneDrive - DOI\mt\cl2022\cl471\from_z3d.h5")
zen_run = m_zen.get_run("471", "sr1_0001", "CL2022")
z_run = zen_run.to_runts()
m_zen.close_mth5()
```

```python
m_geomag = MTH5()
m_geomag.open_mth5(r"c:\Users\jpeacock\OneDrive - DOI\mt\cl2022\cl471\usgs_geomag_frn_xy.h5")
geomag_run = m_geomag.get_run("Fresno", "sp1_001", "USGS-GEOMAG")
g_run = geomag_run.to_runts()
m_geomag.close_mth5()
```

```python
g_slice = g_run.get_slice(z_run.start, end=z_run.end)
```

```python
z_run_calibrated = z_run.calibrate()
```

```python
z_run.plot()
```
