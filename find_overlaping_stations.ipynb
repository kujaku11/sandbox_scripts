```python
import pandas as pd
from pathlib import Path
from mt_metadata.common import MTime
```

```python
survey_summary = pd.read_csv(r"c:\Users\jpeacock\OneDrive - DOI\MTData\CM2025\survey_summary.csv")
```

```python
sdf = survey_summary.sort_values("start").reset_index(drop=True)
```

```python
overlap_dict = {}
overlapping_stations = []
for i, row_i in sdf.iterrows():
    overlap_list = []
    start_i = MTime(time_stamp=row_i["start"])
    end_i = MTime(time_stamp=row_i["end"])
    station_i = row_i["station"]

    for j, row_j in sdf.iterrows():
        if i != j:
            start_j = MTime(time_stamp=row_j["start"])
            end_j = MTime(time_stamp=row_j["end"])
            station_j = row_j["station"]

            if start_j.day == start_i.day:
                start_diff = start_i - start_j
                end_diff = end_i - end_j
                delta = start_diff + end_diff                
                overlap_list.append({"remote": station_j, "overlap": delta})

    find_longest = sorted(overlap_list, key=lambda x: x['overlap'])

    overlapping_stations.append({"local": f"cm{station_i}", "remote": f"cm{find_longest[0]['remote']}"})
```
