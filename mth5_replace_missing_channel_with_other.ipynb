```python
with MTH5() as m:
    m = m.open_mth5(r"c:\Users\jpeacock\OneDrive - DOI\MTData\CM2025\mth5\cm230.h5", mode="a")
    m.remove_run("cm230", "sr1_0001", "cm2025")

    m.channel_summary.summarize()

    rs = m.run_summary
    run_list = []
    for row in rs.sort_values("start").itertuples():
        if row.has_data:
            run_list.append(m.from_reference(row.run_hdf5_reference).to_runts())
    
    combined_run = run_list[0].merge(run_list[1:], new_sample_rate=1)
    combined_run.run_metadata.id = "sr1_0001"
    
    combined_run_group = m.add_run("cm230", "sr1_0001", survey="cm2025")
    combined_run_group.from_runts(combined_run)
    combined_run_group.update_metadata()
    

    
```
